```python
import pandas as pd

df = pd.read_csv('mafs.csv')

df
```

```python
# Counting the succes of doctors.

#fitst getting the total number of entries, divorces, and stay togethers.
counts = df['Decision'].value_counts()
Stayed_Together = counts['Yes']
Broke = counts['No']
```

```python
#creating value counts
experts = df.columns[9:]  # Assuming the expert columns start from the second column
success_rates = {}

for expert in experts:
    success_rate = df[df[expert] == 1]['Decision'].value_counts(normalize=True)['Yes']
    success_rates[expert] = success_rate * 100
```

```python
most_successful_expert = max(success_rates, key=success_rates.get)
```

```python
# Counting the succes of doctors.

#fitst getting the total number of entries, divorces, and stay togethers.
counts = df['Status'].value_counts()
Married = counts['Married']
Divorced = counts['Divorced']
```

```python
#creating value counts
experts = df.columns[9:]  # Assuming the expert columns start from the second column
success_rates = {}

for expert in experts:
    success_rate = df[df[expert] == 1]['Status'].value_counts(normalize=True)['Married']
    success_rates[expert] = success_rate * 100
```

```python
most_successful_expert = max(success_rates, key=success_rates.get)
most_successful_expert
```

```python
df_success_rates = pd.DataFrame(list(success_rates.items()), columns=['Experts', 'Success Rate'])
df_success_rates
```

```python
plt.figure(figsize=(12, 10))
plt.bar(df_success_rates['Experts'], df_success_rates['Success Rate'])
plt.xlabel('Experts')
plt.ylabel('Success Ratio')
plt.xticks(rotation=45) 
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming df_success_rates and df_success_rates2 are two DataFrames containing the data for each expert

# Create positions for the bars
experts = df_success_rates['Experts']
positions = np.arange(len(experts))

# Set the width for each bar
bar_width = 0.35

# Create the figure and axes
plt.figure(figsize=(12, 10))
ax = plt.gca()

# Plot the first set of bars
ax.bar(positions - bar_width/2, df_success_rates['Success Rate'], bar_width, label='Married')

# Plot the second set of bars
ax.bar(positions + bar_width/2, df_success_rates2['Success Rate'], bar_width, label='Said Yes')

# Set the x-axis labels and rotate them for readability
plt.xlabel('Experts')
plt.ylabel('Success Ratio')
plt.xticks(positions, experts, rotation=45)

# Show the legend
plt.legend()

# Show the plot
plt.show()

```

```python
# Create positions for the bars
experts = df_success_rates['Experts']
positions = np.arange(len(experts))

# Set the width for each bar
bar_width = 0.35

# Calculate the difference in success rates between the two dataframes
success_rate_diff = df_success_rates2['Success Rate'] - df_success_rates['Success Rate']

# Create the figure and axes
plt.figure(figsize=(12, 10))
ax = plt.gca()

# Plot the difference in success rates as a bar graph
ax.bar(positions, success_rate_diff, bar_width, label='Success Rate Difference', color='orange')

# Set the x-axis labels and rotate them for readability
plt.xlabel('Experts')
plt.ylabel('False Positive Percentage')
plt.xticks(positions, experts, rotation=45)

# Add text labels above each bar
for i, diff in enumerate(success_rate_diff):
    ax.text(positions[i], diff + 0.01, f"{diff:.2f}", ha='center', va='bottom', fontsize=10)

# Show the plot
plt.show()
```
